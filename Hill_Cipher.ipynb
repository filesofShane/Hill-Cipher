```python
# 1.2

# Importing the numpy library
import numpy as np

# A function to take characters as input and return corresponding numeric values.
def char_to_num(character):
    # This ASCII value of 'A'
    return ord(character) - ord('A')

# A function that is the inverse of char_to_num function
def num_to_char(num):
    return chr(num + ord('A'))

#print(char_to_num(name))
#print(num_to_char(char_to_num(name))) 

# Encryption function
def hill_encrypt(plain_text, key_cipher):
    n = len(key_cipher)
    # This turns our plain_text into capital letters for the 'char_to_num' function
    plain_text = plain_text.upper().replace(' ', '')
    
    if len(plain_text) % n != 0: 
        # This takes the padded plain text and checks if it is divisible by n if it is not it adds 'X's to the end of the text
        # until the padded plain text is divisible by n
        plain_text += 'X' * (n - len(plain_text) % n)
        
    encrypted_text = ''
    
    # Iterating over blocks of length n
    for i in range(0, len(plain_text), n):
        block = plain_text[i:i+n]
        
        # Convert the block characters to their corresponding numeric values
        block_nums = np.array([char_to_num(c) for c in block])
        
        # Encrypt the block using matrix multiplication between key_cipher and block_nums
        encrypted_block_nums = np.dot(key_cipher, block_nums) % 26
        
        # Convert the encrypted block numbers back to characters
        encrypted_block = ''.join([num_to_char(num) for num in encrypted_block_nums])
        
        encrypted_text += encrypted_block
    
    return encrypted_text


name = 'SHANE'
key = [[1, 4, 6], [14, 23, 14], [10, 20, 17]]
name_encrypted = hill_encrypt(name, key)
print('Encrypted Text:', name_encrypted)
```

```python
# 1.3
# Decryption
from sympy import Matrix

A = Matrix(key)
A = A.inv_mod(26)

def hill_decrypt(key_cipher, K_inv):
    n = K_inv.shape[0]
    
    plain_text = []
    
    for i in range(0, len(key_cipher), n):
        block = key_cipher[i:i+n]
        block_nums = np.array([char_to_num(c) for c in block])
        decrypted_block_nums = np.dot(K_inv, block_nums) % 26
        #decrypted_block_nums = decrypted_block_nums.astype(int)
        decrypted_block = ''.join([num_to_char(num) for num in decrypted_block_nums])
        plain_text.append(decrypted_block)
    
    decrypted_text = ''.join(plain_text)
    return decrypted_text



key_inv = A
name_decrypted = hill_decrypt(name_encrypted, key_inv)
# Removing the 'X' character we added
name_decrypted = name_decrypted[:len(name)]
name_decrypted.replace('',' ')

print('Decrypted Text:', name_decrypted)
```
